```python
import pandas as pd
```

```python
# load data

df = pd.read_csv('data/data.csv')
target = 'num_rides'
y = df[target]
X = df.drop(columns=(['Date',target]))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import FeatureUnion
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LinearRegression

def identity(x): return x
# 
num_cols   = ['TAVG', 'TMAX', 'TMIN', 'PRCP', 'SNOW', 'SNWD']
other_cols = ['Holiday', 'weekend']
cat_cols   = []

number_pipe = Pipeline(steps=[
    ('polymer'   , PolynomialFeatures(include_bias=False) ),
    ('scaler'    , StandardScaler()     )
])

binary_pipe = Pipeline(steps=[
    ('ID' , FunctionTransformer(identity , feature_names_out='one-to-one'))
])

category_pipe = Pipeline(steps=[
    ('OneHot' , OneHotEncoder())
])

# Main pipe
pipe = Pipeline(steps=[
    ('pre_prep'   , ColumnTransformer(transformers=[
        ('numerical', number_pipe   , num_cols   ),
        ('binary'   , binary_pipe   , other_cols ),
        ('Cat'      , category_pipe , cat_cols   )
    ])),
    ('Lin_reg' , LinearRegression()
    )
])
```

```python
pipe.fit(X,y)
```

```python
# save main pipeline
import pickle
pkl_filename = 'main_pipeline.pk'
pickle.dump(pipe, open('./app/'+pkl_filename, 'wb'))

# save X column names
pkl_colname = 'columns.pk'
pickle.dump(X.columns, open('./app/'+pkl_colname, 'wb'))
```
